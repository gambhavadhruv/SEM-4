### 01) WAP to demonstrate the use of Pie chart.

```python
x=[10,20,25,40]
y=["food","travel","acomodetion","shopping"]
c=['r','g','y','m']
e=[0,0.2,0,0]
plt.pie(x,labels=y,colors=c,explode=e,shadow=True,autopct="%0.2f%%",textprops={"fontsize":12,"color":"b"},labeldistance=1.3)
plt.show()
```

### 02) WAP to to Plot List random of X, Y Coordinates in Matplotlib.

```python
import random as rd
x=[10,20,30,40]
y=[rd.randint(0,100) for _ in range(0,4)]
plt.plot(x,y)
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
%matplotlib notebook
x=[1,2,3,4,5]
y=[6.3,5.2,4.4,8.1,6.6]
c=['g','b','r','y','m']
l=['1st','2nd','3rd','4th','5th']
# w=[0.5,0.55,0.45,0.4,0.5]
plt.title("Sem wise SPI")
plt.xlabel("sem")
plt.ylabel("spi")
plt.bar(x,y,color=c,edgecolor='b',label=l,alpha=0.6)
plt.show()
```

### 04) WAP to demonstrate the use of Histogram.

```python
x=["Python","Java",".net","C"]
y=[85,70,60,82]
c=['r','g','y','m']
plt.barh(x,y,color=c,edgecolor='r',alpha=0.7)
plt.title("language Population")
plt.xlabel("Rating")
plt.ylabel("Language")
plt.show()
```
